```python
# %% Cell 4 - Clean TotalCharges and drop customerID
# Convert TotalCharges to numeric: there are some blank strings -> coerce
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Option 1: drop small number of rows where TotalCharges is NaN (commonly 11 rows in this dataset)
n_missing_total = df['TotalCharges'].isna().sum()
print("TotalCharges NaNs:", n_missing_total)
if n_missing_total > 0 and n_missing_total < 50:
    df = df.dropna(subset=['TotalCharges']).reset_index(drop=True)
    print(f"Dropped {n_missing_total} rows with missing TotalCharges.")
else:
    # fallback: fill using tenure * MonthlyCharges if many are missing (not the case here)
    df['TotalCharges'] = df['TotalCharges'].fillna(df['tenure'] * df['MonthlyCharges'])

# Drop customerID (not a feature)
if 'customerID' in df.columns:
    df = df.drop('customerID', axis=1)

print("After cleaning:", df.shape)
```

Task **2**

Preprocessing pipeline and train/test split

```python
# %% Cell 7 - Preprocessing pipeline and train/test split
# Prepare X, y
X = df.drop('Churn', axis=1)
y = df['Churn'].map({'No':0, 'Yes':1})  # binary encoding

# Identify feature types
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Categorical:", categorical_features)
print("Numerical:", numerical_features)

# Preprocessor
cat_pipe = Pipeline([
    ('onehot', OneHotEncoder(drop='first', sparse_output=False))
])

num_pipe = Pipeline([('scaler', StandardScaler())])

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', cat_pipe, categorical_features),
        ('num', num_pipe, numerical_features)
    ], remainder='drop'
)

# Train-test split BEFORE SMOTE
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RNG, stratify=y
)
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# Fit-transform training data and transform test data
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)

# Save feature names (useful for interpretation)
oh_names = preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features)
feature_names = list(oh_names) + numerical_features
print(f"Total features after preprocessing: {len(feature_names)}")
```

Smote

```python
# %% Cell 8 - SMOTE balancing (apply to the preprocessed training set)
print("Original training class distribution:", np.bincount(y_train))
sm = SMOTE(random_state=RNG)
X_train_bal, y_train_bal = sm.fit_resample(X_train_prep, y_train)
print("After SMOTE distribution:", np.bincount(y_train_bal))
```

```python
# %% Cell 9 - Baseline Decision Tree (train & evaluate)
dt_base = DecisionTreeClassifier(random_state=RNG, class_weight='balanced')
start = time.time()
dt_base.fit(X_train_bal, y_train_bal)
print("Training time (DT baseline):", time.time()-start)

y_pred = dt_base.predict(X_test_prep)
y_proba = dt_base.predict_proba(X_test_prep)[:,1]

def print_classification(y_true, y_pred, y_proba=None):
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1:", f1_score(y_true, y_pred))
    if y_proba is not None:
        print("ROC AUC:", roc_auc_score(y_true, y_proba))
    print("\nClassification report:\n", classification_report(y_true, y_pred, target_names=['No','Yes']))

print_classification(y_test, y_pred, y_proba)

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred No','Pred Yes'], yticklabels=['Act No','Act Yes'])
plt.title('Confusion Matrix - Baseline DT'); plt.tight_layout(); plt.show()

# Small tree visualization (first 3 levels)
plt.figure(figsize=(14,8))
plot_tree(dt_base, feature_names=feature_names, class_names=['No','Yes'], max_depth=3, filled=True, fontsize=8)
plt.title("Decision Tree (first 3 levels)"); plt.show()
```

```python
# %% Cell 10 - Tuned Decision Tree (small grid for time)
param_grid = {
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini','entropy']
}
dt = DecisionTreeClassifier(random_state=RNG, class_weight='balanced')
gs = GridSearchCV(dt, param_grid, cv=4, scoring='f1', n_jobs=-1, verbose=1)
start = time.time()
gs.fit(X_train_bal, y_train_bal)
print("Grid search time (minutes):", (time.time()-start)/60)
print("Best params:", gs.best_params_, "Best score:", gs.best_score_)

best_dt = gs.best_estimator_
y_pred_t = best_dt.predict(X_test_prep)
y_proba_t = best_dt.predict_proba(X_test_prep)[:,1]
print_classification(y_test, y_pred_t, y_proba_t)
```

```python
# %% Neural Network – Hyperparameter Tuning

print("="*70)
print("SIMPLE HYPERPARAMETER TUNING")
print("="*70)

# Try different hyperparameter combinations
hyperparameter_combinations = [
    {'units_1': 32, 'add_second_layer': True, 'units_2': 16, 'learning_rate': 0.001, 'dropout': 0.2},
    {'units_1': 48, 'add_second_layer': True, 'units_2': 24, 'learning_rate': 0.0005, 'dropout': 0.1},
    {'units_1': 64, 'add_second_layer': False, 'learning_rate': 0.001, 'dropout': 0.3},
    {'units_1': 40, 'add_second_layer': True, 'units_2': 20, 'learning_rate': 0.0001, 'dropout': 0.0},
]

best_accuracy = 0
best_params = None
best_model = None

for i, params in enumerate(hyperparameter_combinations):
    print(f"\n🔧 Testing combination {i+1}/{len(hyperparameter_combinations)}:")
    print(f"  Units: {params.get('units_1')}, "
          f"Second layer: {params.get('add_second_layer', False)}, "
          f"LR: {params.get('learning_rate')}")

    # Build model
    model = Sequential()
    model.add(Dense(params['units_1'], activation='relu', input_dim=input_dim))

    if params.get('add_second_layer', False):
        model.add(Dense(params['units_2'], activation='relu'))

    if params['dropout'] > 0:
        model.add(Dropout(params['dropout']))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=Adam(learning_rate=params['learning_rate']),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    # Train quickly
    history = model.fit(
        X_train_preprocessed, y_train_encoded,
        validation_split=0.2,
        epochs=15,
        batch_size=32,
        verbose=0
    )

    # Evaluate
    _, accuracy = model.evaluate(X_test_preprocessed, y_test_encoded, verbose=0)
    print(f"  Test Accuracy: {accuracy:.4f}")

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_params = params
        best_model = model

print(f"\n🎯 BEST HYPERPARAMETERS FOUND:")
print(best_params)
print(f"Accuracy: {best_accuracy:.4f}")
```
